### 35. Check if two strings are anagrams (simple: check sorted equality).

### 36. Determine the type of triangle using sides: acute, right, obtuse.

```python
# c^2 = a^2 + b^2 Right Angle
# c^2 < a^2 + b^2 Acute Angle
# c^2 > a^2 + b^2 Obtuse Angle

x=3 #side 1
y=4 #side 2
z=5 #side 3

x_sq = x ** 2
y_sq = y ** 2
z_sq = z ** 2

if x + y <= z:
    print( "Not a Triangle")
else:
    if z_sq < x_sq + y_sq:
        print("Acute Triangle")
    elif z_sq == x_sq + y_sq:
        print("Right Triangle")
    else: # z_sq > x_sq + y_sq
        print("Obtuse Triangle")

```

### 40. Check if a number is Armstrong (3-digit).

```python
num40=153
number_str = str(num40)
n = len(number_str)
sum_of_powers = 0
for digit_char in number_str:
    digit = int(digit_char)
    sum_of_powers +=digit** n

if num40 == sum_of_powers:
    print(" Armstrong number.")
else:
    print(" NOT an Armstrong number.")
```

### 42. Check if a number is perfect (sum of factors equals number).

```python
num42=6
sum_of_divisors = 0
if num42 <= 0:
    print("not a positive integer and cannot be perfect.")

else:
    for i in range(1, num42):
        if num42 % i == 0:
            sum_of_divisors += i

    if sum_of_divisors==num42:
        print("Perfect Number!")
    else:
        print("Not a Perfect Number!")
```
